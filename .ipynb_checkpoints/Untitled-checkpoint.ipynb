```python
import numpy as np
```

```python
def nangmean(arr):
    arr = np.asarray(arr)
    inverse = 1 / np.sum(~np.isnan(arr))
    apoio = inverse * np.nansum(np.log(arr))
    nangmean = np.exp(apoio)
    return nangmean
```

```python
nangmean([1,3])
```
